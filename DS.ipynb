```python
import pandas as pd
```

```python
df = pd.read_csv('/content/plant_growth_data.csv')
df
```

```python
data = {
    'Species': ['A', 'B', 'A', 'C', 'B'],
    'Growth_Rate': [1.2, 0.8, 1.5, 1.1, 0.9],
    'Temperature': [20, 25, 22, 18, 24],
    'Sunlight': [8, 9, 10, 7, 8],
    'Biomass': [100, 80, 120, 90, 85]
}
```

```python
df = pd.DataFrame(data)
```

```python
avg_growth_rate = df['Growth_Rate'].mean()
print("Average Growth Rate:", avg_growth_rate)
```

To find the correlation between growth rate and sunlight

```python
correlation = df['Growth_Rate'].corr(df['Sunlight'])
```

```python
print("Correlation between Growth and Sunlight:", correlation)
```
